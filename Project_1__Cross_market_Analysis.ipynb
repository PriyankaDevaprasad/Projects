**Extracting 5 pages of crytocurrency data**

```python
all_records = []
import requests

for i in range(1,6):
  url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&per_page=250&order=market_cap_desc&page={i}&sparkline=False"
  response = requests.get(url)
  print(response)
  if response.status_code == 200:
    data = response.json()
    all_records.extend(data)
```

**Extracting the date and price of top 3 cryptocurrency**

```python
coin_id = ["bitcoin","ethereum","tether"]
coins_hist=[]
for i in coin_id:
  url = f"https://api.coingecko.com/api/v3/coins/{i}/market_chart?vs_currency=usd&days=365"
  response = requests.get(url)
  if response.status_code == 200:
    hist_price=response.json() # pandas dateime and convert this milliseconds and extract only the date
    print(hist_price)
    hist_price=hist_price['prices']
    for j,k in hist_price:
      coins_hist.append({"coin_id":i, "date": pd.to_datetime(j, unit="ms"),"price_usd": k})
print(coins_hist)
```

```python
coins_hist=pd.DataFrame(coins_hist)
coins_hist['date']=coins_hist['date'].dt.date
coins_hist
```

```python
coins_hist.to_csv("Historical_coins.csv",index=False)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import sqlite3
from streamlit_option_menu import option_menu

# Database connection
conn = sqlite3.connect("cmarket.db")

st.title("🪙🛢️📈 Cross-Market Analysis")

with st.sidebar:
    selected = option_menu(
        "Main Menu",
        ["Filters and Data Exploration", "SQL Query runner", "Top 3 Crypto Analysis"],
        icons=['filter', 'search', 'map'],
        menu_icon="cast",
        default_index=0
    )
if selected == "Filters and Data Exploration":

    st.header("Average Price",divider=True)
    st.markdown(''':orange[Select the date range]''')

    import streamlit as st
    col1,col2=st.columns(2)
    with col1:
      start_date = st.date_input("Start Date",width=150,key="start_date",value="2025-01-01")
    with col2:
      end_date = st.date_input("End Date",width=150,key="end_date")

    if start_date and end_date:
      col1, col2=st.columns(2)
      col3,col4 = st.columns(2)
      cp=pd.read_sql(f'''select avg(price_usd) as bitcoin_avg from crypto_prices where coin_id='bitcoin' and date between "{start_date}" and "{end_date}" ''',conn )
      bitcoin=cp['bitcoin_avg'][0]
      op=pd.read_sql(f'''select avg(price) as oil_avg from oilprice where date between "{start_date}" and "{end_date}" ''',conn)
      oil=op['oil_avg'][0]
      sp=pd.read_sql(f'''select avg(close) as sp500 from stock_prices where ticker='^GSPC'and date between "{start_date}" and "{end_date}" ''',conn)
      sp500=sp['sp500'][0]
      nty=pd.read_sql(f'''select avg(close) as Nifty from stock_prices where ticker='^NSEI'and date between "{start_date}" and "{end_date}" ''',conn)
      Nifty=nty['Nifty'][0]
      col1.metric("Average Bitcoin price",round(bitcoin,2) ,border=True,width=1550)
      col2.metric("Average Oil Price",round(oil,2),border=True,width=510)
      col3.metric("Average S&P 500",round(sp500,2),border=True,width=715)
      col4.metric("Average Nifty",round(Nifty,2),border=True,width=715)
    df=pd.read_sql(f'''select c.date,
        c.price_usd as bitcoin_price,o.price AS oil_price,
        s.close as sp500_close
        from crypto_prices c
        join oilprice o on c.date = o.date
        join stock_prices s
        on c.date = s.date
        and s.ticker = '^GSPC'
        where c.coin_id = 'bitcoin'
        and c.date=date('now') or c.date between "{start_date}" and "{end_date}" ''',conn)
    st.subheader("Market Data",divider=True)
    st.dataframe(df)

if selected == "Top 3 Crypto Analysis":
  crypto=st.selectbox("Select a Cryptocurrency",("bitcoin","ethereum"))
  col1,col2=st.columns(2)
  with col1:
      start_date = st.date_input("Start Date",width=150,value="2025-01-01")
  with col2:
      end_date = st.date_input("End Date",width=150)
  df = pd.read_sql(f'''select date,price_usd from crypto_prices where coin_id="{crypto}" and date between "{start_date}" and "{end_date}" ''', conn)
  st.subheader("Filtered Market Data",divider=True)
  st.dataframe(df)
  lc=pd.read_sql(f'''select date,price_usd from crypto_prices where coin_id="{crypto}" and date between "{start_date}" and "{end_date}" order by date''',conn)
  st.subheader(f"{crypto} Price Trend",divider=True)
  st.line_chart(df.set_index("date"))

# ---------------------------------------------------
# QUERIES PAGE
# ---------------------------------------------------
if selected == "SQL Query runner":

    st.header("SQL Query runner")

    option = st.selectbox(
        "Select a Query",
        (
            # Cryptocurrencies
            "Top 3 Cryptocurrencies by Market Cap",
            "Coins With Circulating Supply > 90%",
            "Coins Within 10% of ATH",
            "Average Market Cap Rank (Volume > 1B)",
            "Most Recently Updated Coin",

            # Crypto Prices
            "Highest Bitcoin Price",
            "Average Ethereum Price",
            "Bitcoin January Trend",
            "Coin with highest Average Price",
            "Percentage change in bitcoin price",

            # Oil Queries
            "Highest Oil Price",
            "Average Oil Price Per Year",
            "Covid Crash Price",
            "Lowest Oil Price",
            "Volatality of oilprice",

            # Stock Queries
            "Highest S&P 500 Close",
            "Top 5 Highest Price Difference Days",
            "Monthly Average Close (All Stocks)",
            "Average trading volume in NSEI",


            # Join Queries
            "Bitcoin vs Oil Comparison",
            "Bitcoin vs S&P 500",
            "Ethereum vs NASDAQ",
            "Compare oilprice with bitcoin price",
            "top 3 coins daily price trend vs Nifty",
            "Compare stock prices (^GSPC) with crude oil prices",
            "Correlate Bitcoin closing price with crude oil",
            "Compare NASDAQ (^IXIC) with Ethereum price trends",
            "top 3 crypto coins with stock indices"
        )
    )
 # -------------------------------
    # CRYPTOCURRENCIES
    # -------------------------------

    if option == "Top 3 Cryptocurrencies by Market Cap":
        df = pd.read_sql('''select * from crypto_prices where coin_id='bitcoin' order by price_usd desc limit 1''',conn)
        st.subheader("Top 3 Cryptocurrencies by Market Cap")
        st.dataframe(df)

    if option == "Coins With Circulating Supply > 90%":
        df = pd.read_sql('''select id,name,circulating_supply,total_supply from crypto_currencies where circulating_supply > 0.90 * total_supply''',conn)
        st.subheader("Coins With Circulating Supply > 90%")
        st.dataframe(df)
    if option =="Coins Within 10% of ATH":
      df=pd.read_sql('''select * from crypto_currencies where current_price >= 0.9 * ath''', conn)
      st.subheader("Coins Within 10% of ATH")
      st.dataframe(df)
    if option =="Average Market Cap Rank (Volume > 1B)":
      df=pd.read_sql(''' select name,avg(market_cap_rank) from crypto_currencies where total_volume>1000000000''', conn)
      st.subheader("Average Market Cap Rank (Volume > 1B)")
      st.dataframe(df)
    if option=="Most Recently Updated Coin":
      df=pd.read_sql('''select name,date from crypto_currencies order by date desc limit 10''',conn)
      st.subheader("Most Recently Updated Coin")
      st.dataframe(df)
    if option=="Highest Bitcoin Price":
      df=pd.read_sql('''select * from crypto_prices where coin_id='bitcoin' order by price_usd desc limit 1''',conn)
      st.subheader("Highest Bitcoin Price")
      st.dataframe(df)
    if option =="Average Ethereum Price":
      df=pd.read_sql('''select coin_id,avg(price_usd) from crypto_prices where coin_id='ethereum' ''',conn )
      st.subheader("Average Ethereum Price")
      st.dataframe(df)
    if option =="Bitcoin January Trend":
      df=pd.read_sql('''select * from crypto_prices where coin_id='bitcoin' and date between '2026-01-01' and '2026-01-31' ''',conn)
      st.subheader("Bitcoin January Trend")
      st.dataframe(df)
    if option=="Coin with highest Average Price":
      df=pd.read_sql('''select coin_id,avg(price_usd) from crypto_prices group by coin_id order by avg(price_usd) desc limit 1''',conn)
      st.subheader("Coin with highest Average Price")
      st.dataframe(df)
    if option=="Percentage change in bitcoin price":
      df=pd.read_sql ("""SELECT(SELECT price_usd FROM crypto_prices WHERE coin_id = 'bitcoin' AND date = '2025-09-30') /
(SELECT price_usd FROM crypto_prices WHERE coin_id = 'bitcoin' AND date = '2026-09-30') * 100 AS percent_change;""", conn)
      st.subheader("Percentage change in bitcoin price")
      st.dataframe(df)
    if option=="Highest Oil Price":
      df=pd.read_sql('''select max(price) as highest_oil_price from oilprice where date >=date('now','-5 years')''',conn)
      st.subheader("Highest Oil Price")
      st.dataframe(df)
    if option=="Average Oil Price Per Year":
      df=pd.read_sql('''select strftime('%Y', date) AS year,avg(price) as avg_oil_price from oilprice group by year order by year''',conn)
      st.subheader("Average Oil Price Per Year")
      st.dataframe(df)
    if option == "Covid Crash Price":
      df=pd.read_sql('''select strftime('%Y-%m',date) as month,avg(price) from oilprice group by month having month between '2020-03' and '2020-04' ''',conn)
      st.subheader("Covid Crash Price")
      st.dataframe(df)
    if option =="Lowest Oil Price":
      df=pd.read_sql('''select min(price) as lowest_oil_price from oilprice where date >=date('now','-10 years')''',conn)
      st.subheader("Lowest Oil Price")
      st.dataframe(df)
    if option=="Volatality of oilprice":
      df=pd.read_sql('''select strftime('%Y', date) AS year,max(price)-min(price) as volatility from oilprice group by year order by year''',conn)
      st.subheader("Volatality of oilprice")
      st.dataframe(df)
    if option=="Highest S&P 500 Close":
      df=pd.read_sql('''select * from stock_prices where ticker='^GSPC' ''',conn)
      st.subheader("Highest S&P 500 Close")
      st.dataframe(df)
    if option=="Top 5 Highest Price Difference Days":
      df=pd.read_sql('''select max(close) from stock_prices where ticker='^IXIC' ''',conn)
      st.subheader("Top 5 Highest Price Difference Days")
      st.dataframe(df)
    if option=="Monthly Average Close (All Stocks)":
      df=pd.read_sql(''' select date,high-low as high_price_difference from stock_prices where ticker='^GSPC' order by high_price_difference desc limit 5''',conn)
      st.subheader("Monthly Average Close (All Stocks)")
      st.dataframe(df)
    if option=="Average trading volume in NSEI":
      df=pd.read_sql('''select avg(volume) as avg_voloume_2024,ticker from stock_prices where ticker='^NSEI' and strftime('%Y',date)='2024' ''',conn)
      st.subheader("Average trading volume in NSEI")
      st.dataframe(df)
    if option=="Bitcoin vs Oil Comparison":
      df=pd.read_sql('''select c.date,c.price_usd as bitcoin_price,s.close as stock_closeprice from crypto_prices c join stock_prices s on c.date = s.date where c.coin_id='bitcoin' and s.ticker = '^GSPC' ''',conn)
      st.subheader("Bitcoin vs Oil Comparison")
      st.dataframe(df)
    if option=="Bitcoin vs S&P 500":
      df=pd.read_sql('''select c.date,c.price_usd as bitcoin_price,s.close as stock_closeprice from crypto_prices c join stock_prices s on c.date = s.date where c.coin_id='bitcoin' and s.ticker = '^GSPC' ''',conn)
      st.subheader("Bitcoin vs S&P 500")
      st.dataframe(df)
    if option=="Ethereum vs NASDAQ":
      df=pd.read_sql('''select c.date,c.price_usd as ethereum_price,s.close as NASDAQ_price from crypto_prices c join stock_prices s on c.date = s.date where c.coin_id = 'ethereum' and s.ticker = '^IXIC' and strftime('%Y', c.date) = '2025' ''',conn)
      st.subheader("Ethereum vs NASDAQ")
      st.dataframe(df)
    if option=="Compare oilprice with bitcoin price":
      df=pd.read_sql('''select c.price as oilprice,s.price_usd as bitcoin_price from oilprice c join crypto_prices s on c.date=s.date where s.coin_id='bitcoin' order by oilprice desc limit 5''',conn)
      st.subheader("Compare oilprice with bitcoin price")
      st.dataframe(df)
    if option=="top 3 coins daily price trend vs Nifty":
      df=pd.read_sql('''select c.date,c.coin_id,c.price_usd,s.close as NSEI_price from crypto_prices c join stock_prices s on c.date=s.date where ticker='^NSEI' order by c.date desc''',conn)
      st.subheader("top 3 coins daily price trend vs Nifty")
      st.dataframe(df)
    if option=="Compare stock prices (^GSPC) with crude oil prices":
      df=pd.read_sql('''select c.date,c.close as stock_price,s.price as oilprice from stock_prices c join oilprice s on c.date=s.date where ticker='^GSPC' ''',conn)
      st.subheader("Compare stock prices (^GSPC) with crude oil prices")
      st.dataframe(df)
    if option=="Correlate Bitcoin closing price with crude oil":
      df=pd.read_sql('''select c.date,c.price_usd as bitcoin_price,s.price as oilprice from crypto_prices c join oilprice s on c.date=s.date where coin_id='bitcoin' ''',conn)
      st.subheader("Correlate Bitcoin closing price with crude oil")
      st.dataframe(df)
    if option=="Compare NASDAQ (^IXIC) with Ethereum price trends":
      df=pd.read_sql('''select c.close,s.price_usd from stock_prices c join crypto_prices s on c.date = s.date where ticker='^IXIC' and coin_id='ethereum' ''',conn)
      st.subheader("Compare NASDAQ (^IXIC) with Ethereum price trends")
      st.dataframe(df)
    if option=="top 3 crypto coins with stock indices":
      df=pd.read_sql('''select c.date,c.coin_id,s.close,s.ticker from crypto_prices c join stock_prices s on c.date=s.date where strftime('%Y',c.date)='2025' ''',conn)
      st.subheader("top 3 crypto coins with stock indices")
      st.dataframe(df)
```
